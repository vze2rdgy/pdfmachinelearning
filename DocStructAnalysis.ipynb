```python
def getDocProperties(lines):
    docProperties = {"longestsent" : 0}
    lineproperties = {}
    for line in lines:
        if line:
            linemeta = lineproperties.get(line)
            if not linemeta:
                linemeta = {}
                lineproperties[line] = linemeta
            if line[0] != ' ':
                linemeta['startcol'] = 0
            else:
                for i, c in enumerate(line):
                    if c != ' ':
                        linemeta['startcol'] = i
                        break
            linemeta['fulllength'] = len(line)
            linemeta['actuallenght'] = len(line.strip())

            longestsent = docProperties.get("longestsent")
            l = longestsent
            l = max(l, len(line))
            docProperties["longestsent"] = l
    return (lineproperties, docProperties)
```
